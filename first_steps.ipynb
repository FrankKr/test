```python
# first import some necessary packages/functions
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# load an example data set (random first 200 lines)
# in the csv no column names are specified so we provide them
zowndata = pd.read_csv('testdata.csv', header = None, index_col = False, 
                           names = ['var', 'var2', 'unit', 'value', 'value_float', 'unknown_var', 'datetime'])
```

```python
# convert the datetime column to the right format
zowndata['datetime']  = pd.to_datetime(zowndata['datetime'], format = '%Y-%m-%d %H:%M:%S.%f')

# order by datetime
zowndata.sort_values(by = 'datetime', inplace = True)

# make a simple plot
plt.plot(zowndata['datetime'], zowndata['value_float'])
```
